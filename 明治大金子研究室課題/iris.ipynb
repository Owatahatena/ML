```python
from sklearn.datasets import load_iris
```

```python
import numpy as np
import pandas as pd
```

```python
iris = load_iris()
```

```python
iris
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df
```

```python
"""
課題1: iris.csvを読み込み各変数の最大値・最小値・平均値・中央値・分散・標準偏差を求めbasic_statistics.csvに保存するプログラムを作成せよ。
自分なりの方法で各統計量があっているか確認せよ
"""
```

```python
df.describe()
```

```python
dfv = pd.DataFrame(df.var())
dfv = dfv.T
dfv.index   = ['var']
dfv
```

```python
dfm = pd.DataFrame(df.median())
dfm = dfm.T
dfm.index   = ['median']
dfm
```

```python
pd.concat([df.describe(),dfv,dfm])
```

```python
"""
課題2: iris.csvを読み込みすべての変数間の共分散と相関係数を求めそれぞれcovariance.csv, correlationcoefficient.csvに保存するプログラムを作成せよ。
自分なりの方法で各統計量があっているか確認せよ。
"""
```

```python
df.corr() #相関係数
```

```python
df.cov() #共分散
```

```python
"""
課題3: まずは標準化(オートスケーリング)について調べよ。次に、iris.csvを読み込みオートスケーリングを行ってautoscaled_iris.csvに保存するプログラムを作成せよ。
自分なりの方法で結果があっているか確認せよ。
"""
```

```python
from sklearn import preprocessing
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
np_scaled = min_max_scaler.fit_transform(df)
df_normalized = pd.DataFrame(np_scaled)
df_normalized
```

```python
"""
課題6: iris.csvを読み込みオートスケーリングをしてから主成分分析を行いデータの可視化をせよ。可視化の様子からいえることを考察せよ。
"""
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(df_normalized)
```

```python
X_pca= pca.transform(df_normalized)
```
